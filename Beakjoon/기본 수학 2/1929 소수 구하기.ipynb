```python
# 10:18 ~ 11:31

# 에라토스테네스의 체 알고리즘을 사용하는 문제이다.
# 먼저 구할 만큼의 배열을 선언한 후, 소수가 아닌 숫자들을 하나씩 지워가는 방식이 에라토스테네스의 체 알고리즘이다.
# 1 <= n <= m <= 1,000,000

import math

n, m = map(int, input().split())

# 소수가 반드시 하나는 있으므로 n = 2로 설정해주어도 아무 문제 없다. 최소의 경우에 n = 2, m = 2가 되어 반복문이 실행될 것이다.
if (n == 1) :
    n = 2
    
# 배열을 생성하고 초기화해준다.
prime_arr = []
for i in range(m + 1) :
    prime_arr.append(i)
    
for i in range (n) :
    prime_arr[i] = 0

prime_arr[0] = 0
prime_arr[1] = 0

    
for i in range (2, m + 1) :
    
    for j in range (i + i, m + 1, i)  :
        if (prime_arr[j] == 0) :
            continue
        prime_arr[j] = 0
    
for i in range(len(prime_arr)) :
    if (prime_arr[i] != 0) :
        print(prime_arr[i])
```
