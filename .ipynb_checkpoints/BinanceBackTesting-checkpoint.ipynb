```python
# 단순 변동성 돌파 전략

import ccxt
import pprint
import numpy as np
import pandas as pd
import datetime

binance = ccxt.binance({
    'apiKey' : 'vDiLBxzA2VPlSuBLYXiDF47mScyldTANFIdysXlZYbVtcM9MA7g65F0EKPwG1w9k',
    'secret' : '3gs0sLJ0HnLsHBmcnqEKH46nEklakikZubDefJEbK8wOusHA8XiDtYj0Xw6LrnLY',
})
# 기본 변수 설정 **********************************************

ticker = "BTC/USDT"
k = 0.5
btc_ohlcv = binance.fetch_ohlcv(ticker, "1d")
df = pd.DataFrame(btc_ohlcv, columns=['datetime', 'open', 'high', 'low', 'close', 'volume'])

# ********************************************************************************

#  변동폭 * k 계산, (고가 - 저가) * k값
df['range'] = (df['high'] -df['low']) * 0.5

# target(매수가), range 컬럼을 한 칸씩 밑으로 내림
df['target'] = df['open'] + df['range'].shift(1)


# ror(수익률), np.where(조건문, 참일때 값, 거짓일때 값)
df['ror'] = np.where(df['high'] > df['target'],
                    df['close'] / df['target'], 1)

# 누적 곱 계산(cumprod) => 누적 수익률
df['hpr'] = df['ror'].cumprod()


# Draw Down 계산 (누적 최대 값과 현재 hpr 차이 / 누적 최대값 * 100)
df['dd'] = (df['hpr'].cummax() - df['hpr']) / df['hpr'].cummax() * 100


# MDD 계산
print("MDD(%): ", df['dd'].max())

# 엑셀로 출력 
df.to_excel(excel_writer='normalDD.xlsx')






# df = pd.DataFrame(btc_ohlcv, columns=['datetime', 'open', 'high', 'low', 'close', 'volume'])
# df['datetime'] = pd.to_datetime(df['datetime'], unit='ms')
# df.set_index('datetime', inplace=True)
# print(df)
```
