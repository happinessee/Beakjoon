```python
def solve (a) :
    if a<100 :
        return 1
    else :
        a = list(str(a))
        if(int(a[0]) - int(a[1]) == int(a[1]) - int(a[2])) :
            return 1
        elif a == 1000 :
            return 0
        return 0

n=int(input())
count = 0
for i in range(1, n + 1) :
    count += solve(a = i)
print(count)
```
